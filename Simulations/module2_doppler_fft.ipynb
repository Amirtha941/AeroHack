```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated wingbeat: periodic motion + noise
fs = 1000  # sampling rate (Hz)
t = np.linspace(0, 1, fs)
wingbeat_freq = 5  # Hz
signal = np.sin(2 * np.pi * wingbeat_freq * t) + 0.5 * np.random.randn(len(t))

# FFT analysis
fft_signal = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(t), 1/fs)

# Plot positive frequency spectrum
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:fs//2], np.abs(fft_signal[:fs//2]))
plt.title("Simulated Doppler Wing Flap Detection (FFT)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.savefig("module2_doppler_fft.png")
plt.show()
```
