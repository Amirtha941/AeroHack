```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated straight flight path (X axis)
t = np.linspace(0, 10, 100)
x = t
y_straight = np.zeros_like(t)

# Simulated distorted path (with wobble from IMU input)
wobble_amplitude = 0.5
wobble_frequency = 2
y_wobble = wobble_amplitude * np.sin(2 * np.pi * wobble_frequency * t)

# Plot both paths
plt.figure(figsize=(10, 5))
plt.plot(x, y_straight, label='Straight Path (No Distortion)', linestyle='--')
plt.plot(x, y_wobble, label='Distorted Path (IMU Randomness)', color='orange')
plt.xlabel("Forward Distance (m)")
plt.ylabel("Lateral Deviation (m)")
plt.title("Drone Path: With vs Without IMU-Based Distortion")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("module1_path_distortion.png")
plt.show()
```
