```python
import matplotlib.pyplot as plt
import numpy as np
# Simulate IR pulse: 1 = ON, 0 = OFF
duration = 2  # seconds
fs = 1000
t = np.linspace(0, duration, fs*duration)
pulse_freq = 2  # 2 Hz pulse
pwm_signal = (np.sin(2 * np.pi * pulse_freq * t) > 0).astype(int)

plt.figure(figsize=(10, 2))
plt.plot(t, pwm_signal, drawstyle='steps-post')
plt.title("Simulated IR Beacon Pulse (PWM Signal)")
plt.xlabel("Time (s)")
plt.ylabel("IR LED State")
plt.yticks([0,1], ["OFF", "ON"])
plt.grid(True)
plt.tight_layout()
plt.savefig("module4_ir_pulse.png")
plt.show()
```
